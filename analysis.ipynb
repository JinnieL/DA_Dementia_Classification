### 데이터셋 설명       
60세 ~ 96세 사이의 150명의 피험자를 대상으로 한 데이터.
각 피험자(Subject ID)는 1년 이상의 간격을 두고 2회 이상 방문(MRI ID & Visit).                           
피험자는 모두 오른손잡이.           
피험자 72명은 연구기간 내내 비치매로 분류됨.            
64명은 최초 방문시 치매로 분류되어 이후 스캔에서도 치매로 분류됨.           
14명은 최초 방문시 비치매였으나 이후 치매로 분류됨.

컬럼 설명
- Subject ID
- MRI ID
- Group : Nondemented / Demented / Converted 
- Visit : 방문 횟수
- MR Delay 
- M/F : 성별
- Hand : 주요 사용 손
- Age : 나이
- EDUC : 교육 년수
- SES : Social Economic Status as assessed by the Hollingshead Index of Social Position and classified into categories from 1 (highest status) to 5 (lowest status) 사회 경제적 지위

- MMSE : Mini-Mental State Examination score (0 = worst ~ 30 = best)
    - 중증(9점 이하)                
    - 중등도(10~18점)               
    - 경도(19~23점)             
    - 24점(30점 만점) 이상의 점수는 정상적인 인지   
- CDR : Clinical Dementia Rating 
    - 0 = 정상
    - 0.5 = 매우 경도 치매
    - 1 = 경도 치매
    - 2 = 중증 치매
    - 3 = 심각한 치매
- eTIV : Estimated total intracranial volume, mm3 (추정된 총 두개 부피)
- nWBV : Normalized whole-brain volume, expressed as a percent of all voxels in the atlas-masked image that are labeled as gray or white matter by the automated tissue segmentation process
- ASF : Atlas scaling factor (unitless). Computed scaling factor that transforms native-space brain and skull to the atlas target (i.e., the determinant of the transform matrix)

#### 참여자 전체 수 & 진단 그룹별 수 확인

```python
# 전체 참여자 수 확인
all = dementia['Subject ID'].nunique()
all
```

```python
# 참여자 중 첫 방문시 치매 X → 추후 방문시 치매로 전환된 참여자 수 확인
converted = dementia[dementia['Group'] == 'Converted']['Subject ID'].nunique()
converted
```

```python
# 연구 기간 내내 치매였던 참여자 수 확인
demented = dementia[dementia['Group'] == 'Demented']['Subject ID'].nunique()
demented
```

```python
# 연구 기간 내내 정상이었던 참여자 수 확인
nondemented = dementia[dementia['Group'] == 'Nondemented']['Subject ID'].nunique()
nondemented
```

```python
converted + demented + nondemented == all
```
